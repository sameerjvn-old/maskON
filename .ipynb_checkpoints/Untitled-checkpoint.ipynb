```python
from fastai.vision.widgets import *
import torchvision.transforms as T
```

```python
URLHolder = widgets.Text(
    value='https://upload.wikimedia.org/wikipedia/commons/c/c3/COVID-19_%28Coronavirus%29_Girl_in_mask.jpg',
    placeholder='Type something',
    description='URL:',
    disabled=False
)
```

```python
display(VBox([widgets.Label('Upload a face! (and hit enter)'), URLHolder]))
```

```python
print(URLHolder.value)
```

```python
url = URLHolder.value
```

```python
from PIL import Image
import requests
import matplotlib.pyplot as plt

response = requests.get(url, stream=True)
img = Image.open(response.raw)

plt.imshow(img)
plt.show()
```
